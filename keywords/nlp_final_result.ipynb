```python
import json
```

```python
with open('results.json') as json_file:
    data = json.load(json_file)
```

```python
# data: preporcessed restaurants avg stars, and top 10 key words
data
```

```python
# recommended_keywords = ['good', 'great','service', 'pizza', 'nice', 'friendly', 'chicken', 'amazing', 'fries', 'sandwich', 'mexican', 'breakfast', 'new']
```

```python
# input: a list of keywords, a number of recommended restaurants
# output: a list of n recommended restaurant ids
def get_recommends(keywords, n = 10):
    res = []
    temp = {}
    for i in data.keys():
        score = data[i]['star']/10
        for j in keywords:
            if j in data[i]['keywords']:
                score += 1
       
        temp[i] = score
    sorted_temp = sorted(temp.items(), key=lambda x: x[1], reverse = True)  
    for k in range(n):
        res.append(sorted_temp[k][0])
    return res
```

```python
recommend_ids = get_recommends(['good', 'pizza'])
```
